# Data quality checks

## Finding Duplicates

```python
Lead_dup = Lead.copy()

# Checking for duplicates and dropping the entire duplicate row if any
Lead_dup.drop_duplicates(subset=None, inplace=True)
```

```python
Lead_dup.shape
```

### Insights:
- Zero duplicates were found in the data set after using drop command & comparing the shape of the data frames.

## Finding null values:

```python
# finding nul values in columns of the dataset
round((Lead.isnull().sum()/len(Lead))*100,2).sort_values(ascending=False)
```

```python
# finding null values in rows of the data set
round((Lead.isnull().sum(axis=1)/len(Lead))*100,2).sort_values(ascending=False)
```
